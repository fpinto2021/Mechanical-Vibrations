```python
import numpy as np
import matplotlib.pyplot as plt

# Given
s = 27
n = s + 5
L = 1

# Defining x
x = np.linspace(0, L, 500)

# Computing mode shape
phi_n = np.sin(n * np.pi * x / L)

# Plotting results
plt.figure(figsize=(8, 5))
plt.plot(x, phi_n, color='pink', label="Mode Shape")
plt.xlabel("Position along the beam (x/L)")
plt.ylabel("Mode Shape Amplitude")
plt.title("Mode Shape")
plt.axhline(0, color='black', linewidth=0.8, linestyle="--")
plt.legend()
plt.grid()
plt.show()
```
