```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.integrate import odeint

# Given=
m = 27
k = 1
t1 = 4
F0 = 20

# Natural frequency
wn = np.sqrt(k / m)

# Defining force function
def F(t):
    return np.where(t <= t1, (F0 * t) / t1, F0)

# Defining system of equations
def system(y, t, wn, t1, F0):
    x, v = y
    force = F(t)
    dxdt = v
    dvdt = (force - k * x) / m
    return [dxdt, dvdt]

# Initial conditions
y0 = [0, 0]

# Time range
t = np.linspace(0, 20, 1000)

# Solve ODE
solution = odeint(system, y0, t, args=(wn, t1, F0))
x_response = solution[:, 0]

# Plotting results
plt.figure(figsize=(8, 5))
plt.plot(t, x_response, label="Displacement", color='pink')
plt.axvline(x=t1, color='r', linestyle='--', label="$t_1$")
plt.xlabel("Time (s)")
plt.ylabel("Displacement (m)")
plt.title("Undamped System")
plt.legend()
plt.grid()
plt.show()
```
